```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
df =pd.read_csv("../data/cleaned_owid-covid-data.csv")
df.head()
```

```python
# show bar graph for continents and total deaths
plt.bar(df["continent"],df['new_deaths_per_million'])
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(x='total_cases', y='continent', data=df, hue="date", style='date')
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```
